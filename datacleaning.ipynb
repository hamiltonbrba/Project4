## 5. Creating a Binary Dataset for Diabetes vs. No Diabetes
To create a binary classification dataset distinguishing diabetes from no diabetes, we can handle prediabetic individuals in one of three ways: grouping them with diabetics, grouping them with non-diabetics, or excluding them entirely. For this analysis, I propose grouping prediabetics with diabetics. A prediabetes diagnosis indicates a heightened risk of diabetes, aligning with our focus on understanding risk.

We can also balance the dataset into a 50-50 split between non-diabetics and the combined group of diabetics and prediabetics. There are enough records available to accomplish this randomly while maintaining predictive value. Specifically, we can combine the 0 prediabetic cases and 6,581 diabetic cases to form a group of 6,581 individuals at diabetes risk. This can be matched with 6,581 randomly selected non-diabetic cases from the 255,008 available.

It's important to note that, as this dataset is survey-based, some individuals may not yet have been diagnosed as prediabetic or diabetic, meaning undiagnosed cases could exist.

```python
#Copy old table to new one.
brfss_binary = brfss
#Change the diabetics 2 to a 1 and pre-diabetics 1 to a 0, so that we have 0 meaning non-diabetic and pre-diabetic and 1 meaning diabetic.
brfss_binary['Diabetes_012'] = brfss_binary['Diabetes_012'].replace({1:0})
brfss_binary['Diabetes_012'] = brfss_binary['Diabetes_012'].replace({2:1})

#Change the column name to Diabetes_binary
brfss_binary = brfss_binary.rename(columns = {'Diabetes_012': 'Diabetes_binary'})
brfss_binary.Diabetes_binary.unique()
```

```python
#Show the change
brfss_binary.head()
```

```python
#show class sizes
brfss_binary.groupby(['Diabetes_binary']).size()
```

```python
# Separate the 0 (Non-diabetes) and 1 (Diabetes risk)
# Get the 1s (Diabetes risk group)
brfss_5050_1 = brfss_binary[brfss_binary['Diabetes_binary'] == 1]

# Get the 0s (Non-diabetes group)
brfss_5050_0 = brfss_binary[brfss_binary['Diabetes_binary'] == 0]

# Randomly select the same number of cases as the diabetes risk group (to balance classes)
num_diabetes_cases = len(brfss_5050_1)  # Get the number of cases in the diabetes risk group
brfss_5050_0_rand = brfss_5050_0.sample(n=num_diabetes_cases, random_state=42)

# Combine the sampled 0s and all 1s into a balanced dataset
brfss_5050 = pd.concat([brfss_5050_0_rand, brfss_5050_1], ignore_index=True)

# Verify the new balanced dataset
print(brfss_5050['Diabetes_binary'].value_counts())
```

```python
#Check that it worked. Now we have a dataset of 79,954 rows that is equally balanced with 50% 1 and 50% 0 for the target variable Diabetes_binary
brfss_5050.head()
```

```python
# Reclassify the Diabetes_binary column
brfss_binary['Diabetes_binary'] = brfss_binary['Diabetes_binary'].replace({
    0: 0,  # No diabetes or only during pregnancy
    3: 0,  # No
    4: 1,  # Pre-diabetes or borderline diabetes
    1: 2,  # Yes
    2: 2   # Yes, but female told only during pregnancy
})

# Verify the unique values and their counts
print(brfss_binary['Diabetes_binary'].value_counts())
```

```python
# Separate the 0s (No diabetes), 1s (Pre-diabetes), and 2s (Diabetes)
brfss_0 = brfss_binary[brfss_binary['Diabetes_binary'] == 0]
brfss_1 = brfss_binary[brfss_binary['Diabetes_binary'] == 1]
brfss_2 = brfss_binary[brfss_binary['Diabetes_binary'] == 2]

# Balance the dataset (if needed, e.g., 50-50 split for modeling)
# Here, balancing the non-diabetes (0) with the diabetes-risk group (1 and 2 combined)
diabetes_risk_group = pd.concat([brfss_1, brfss_2])
num_diabetes_cases = len(diabetes_risk_group)

# Randomly sample from the non-diabetes group to match the size of the diabetes-risk group
brfss_0_sample = brfss_0.sample(n=num_diabetes_cases, random_state=42)

# Combine into a balanced dataset
brfss_5050 = pd.concat([brfss_0_sample, diabetes_risk_group], ignore_index=True)

# Verify the balanced dataset
print(brfss_5050['Diabetes_binary'].value_counts())
```

```python
#Check that it worked. Now we have a dataset of 79,954 rows that is equally balanced with 50% 1 and 50% 0 for the target variable Diabetes_binary
brfss_5050.head()
```
